```python
b. In case amount spend is more than the limit, replace it with that 50% of the Customer’s Limit.
```

```python
Customer_Spend_Trans= pd.merge(right= Customer_Acquisition, left= Spend, on= 'Customer')
Customer_Spend_Trans
```

```python
Customer_Repay_Trans= pd.merge(right= Customer_Acquisition, left= Repayment, on= 'Customer')
```

```python
Customer_Repay_Trans['Amount']= np.where(Customer_Repay_Trans.Amount> Customer_Spend_Trans.Limit,Customer_Spend_Trans.Limit,Customer_Spend_Trans.Amount)
```

```python
Customer_Spend_Amount= Customer_Spend_Trans.groupby(['Customer','Month'])[['Amount']].mean().reset_index()
Customer_Spend_Amount['Spend_Month']= Customer_Spend_Amount['Month'].apply(lambda x:x.month)
Avg_Montly_Spend= Customer_Spend_Amount.groupby(['Customer','Spend_Month'])[['Amount']].mean().reset_index() 

Avg_Montly_Spend
```

```python
2.d. What is the average monthly repayment by customers?
```

```python
Customer_Repay_Amount= Customer_Repay_Trans.groupby(['Customer','Month'])[['Amount']].mean().reset_index()
Customer_Repay_Amount['Repay_Month']= Customer_Repay_Amount['Month'].apply(lambda x:x.month)
Avg_Montly_Repayment= Customer_Repay_Amount.groupby(['Customer','Repay_Month'])[['Amount']].mean().reset_index() 

Avg_Montly_Repayment
```

```python
Avg_Monthly_Customers= pd.merge(left= Avg_Montly_Spend, right= Avg_Montly_Repayment, left_on= ['Customer','Spend_Month'], right_on= ['Customer','Repay_Month'] )
Avg_Monthly_Customers
```
